```python
#importing the module newspaper
import newspaper

#displaying urls examples for user input
print('''
Articles Urls:

https://www.theguardian.com/world/2023/jun/13/call-overhaul-out-of-date-uk-abortion-laws-woman-jailed

https://www.theguardian.com/uk-news/2023/jun/12/three-britons-confirmed-dead-in-egypt-boat-fire

https://www.theguardian.com/environment/2023/jun/12/uk-heatwave-prompts-order-to-fire-up-coal-plant-to-meet-aircon-demand

      ''')
#defining a function to extract the content from the url address
def read_articles(urls_address):
    articles = []

    for url_address in urls_address:
      try:
        article = newspaper.Article(url_address)
        article.download()
        article.parse()
        title = article.title
        text = article.text
        articles.append({"title": title, "text": text})

#error handling in case the user input a wrong datatype or mistake in the url
      except Exception as e:
        print(f"Invalid URL: {url_address}")
        print(f"Error!: {str(e)}")
        print()
    return articles

#asking the user to input the number of articles urls they wish to extract
number_urls = int(input("Enter the number of urls: "))
urls_address = []

#asking the user to input the url addresses the number of times they put in the first user input
for i in range (number_urls):
    url_input = input("Enter url address: ")
    urls_address.append(url_input)

#display the extracted content
content = read_articles(urls_address)
for article in content:
    print("\nTitle:", article["title"])
    print("\nText:", article["text"])
    print("*****************************************************")
```

2. The role of the Newspaper3k library is to make it easier and less bulky to process the articles' extraction and parsing.

3. To handle issues such as invalid urls, I would use error handling and display an error indicating the user that the url is invalid.
